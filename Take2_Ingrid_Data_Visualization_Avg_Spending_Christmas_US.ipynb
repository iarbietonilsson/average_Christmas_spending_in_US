* ### Fill in area under curve of lineplot
   * ### *More festive !*

```python
# set figure size very wide
fig, ax = plt.subplots(figsize=(20,8))

# plot lineplot
sns.lineplot(data=df, x='Year',y='Christmas_spend_in_US', color = '#34c284', marker='o')

# add title
plt.title('Average Spending for Christmas Gifts per US Household (1999 - 2018 )', fontsize=20, fontweight='bold');

# set x axis and y axis limits
plt.xlim(left=1998, right=2019)
plt.ylim(bottom=0,top=1100)

# add x axis label, y axis label, and title
plt.xlabel('Year', fontsize=20, fontweight='bold');
plt.ylabel('Christmas Spending in the US ($)', fontsize=20, fontweight='bold');

# set x axis tick values for each year
ax.set_xticks([1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]);

# add marker values per datapoint
for Year, Christmas_spend_in_US in zip(df['Year'], df['Christmas_spend_in_US']):
    ax.annotate(f'${Christmas_spend_in_US}', xy=(Year,Christmas_spend_in_US))

# fill area under line, green color for Christmas color as well as the color of money
plt.fill_between(df.Year.values, df.Christmas_spend_in_US.values, color='#7FFFD4');
```

# **Final Visualization**

---



```python
# set figure size very wide
fig, ax = plt.subplots(figsize=(20,8))

# plot lineplot
sns.lineplot(data=df, x='Year',y='Christmas_spend_in_US', color = '#34c284', marker='o')

# add title
plt.title('Average Spending for Christmas Gifts per US Household (1999 - 2018 )', fontsize=20, fontweight='bold');

# set x axis and y axis limits
plt.xlim(left=1998, right=2019)
plt.ylim(bottom=0,top=1100)

# add x axis label, y axis label, and title
plt.xlabel('Year', fontsize=20, fontweight='bold');
plt.ylabel('Christmas Spending in the US ($)', fontsize=20, fontweight='bold');

# set x axis tick values for each year
ax.set_xticks([1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]);

# add marker values per datapoint
for Year, Christmas_spend_in_US in zip(df['Year'], df['Christmas_spend_in_US']):
    ax.annotate(f'${Christmas_spend_in_US}', xy=(Year,Christmas_spend_in_US))

# fill area under line, green color for Christmas color as well as the color of money
plt.fill_between(df.Year.values, df.Christmas_spend_in_US.values, color='#7FFFD4');
```

#### This line plot shows the average spending for Christmas gifts per US household from 1999 through 2018. 
* #### Including the Title makes it clear to audience what the visualization is plotting.
* #### Labeling the x-axis & y-axis, as well as adjusting the tick interval makes it more visually appealing.
* #### Including $ Dollar signs conveys that we're talking about currency.
* #### Filling area under curve is also for visual effect, in a merry Christmas color!
